This is a notebook to tell you how to use the local mongodb database

1.The way to get access to the database

```python
import pymongo
import pandas as pd
```

```python
#Note right now, everything is saved on database named gdelt_database and 
#collection named current_data_collection
client = pymongo.MongoClient('localhost', 27017) # This is the way to get access to the mongodb
db = client['gdelt_database'] #This is the way to the specific database
collection = db['current_data_collection'] #This is the way to specific collection
```

Collection is a pymongo object and you can get access to the collection by some methods

Get one data

```python
# Get one document
# it is important to run one data to see which kind of data you have and see all variables in the db
collection.find_one()
# We can the data in the mongodb is dictionary and the collection is a list of dictionary
```

Get the whole dataset

```python
whole_data_set = list(collection.find())
#This is the syntax to get the whole dataset. Because it contains 6 GB data, reading can be slow
# I do not run it and do not run it unless it is necessary
```

```python
#By using limit, it only gets 500 documents in the db but the syntax is same
part_of_whole_data = list(collection.find().limit(500))
```

Transfer the list of dictionary into pd dataframe

```python
part_data_dataframe = pd.DataFrame(part_of_whole_data)
```

```python
#Have a look at the dataframe
part_data_dataframe.head()
```

Data query in mongodb

```python
#Data query in mongodb is straightforward
# This is an example to query the data in a specific data and month
oneday_data = list(collection.find({'SQLDATE':20150225}))
```

```python
def data_query_plotical_risk(input_date):
    input_date = int(input_date)
    
    #Connect to the database
    client = pymongo.MongoClient('localhost', 27017) # This is the way to get access to the mongodb
    db = client['gdelt_database'] #This is the way to the specific database
    collection = db['current_data_collection'] #This is the way to specific collection
    
    oneday_data = list(collection.find({'SQLDATE':input_date}))
    oneday_dataframe = pd.DataFrame(oneday_data)
    return oneday_dataframe
```

```python
#You can run the function to query the daily data
test_data_frame = data_query_plotical_risk('20150310')
test_data_frame.head()
```
