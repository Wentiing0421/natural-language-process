```python
import pandas as pd
import numpy as np
```

```python
#list = [result_2014,result_2015,result_2016,result_2017,result_2018,result_2019]
arg2014 = pd.read_csv('result_2014.csv',index_col=False)
arg2014 = arg2014.drop(arg2014.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 

arg2015 = pd.read_csv('result_2015.csv',index_col=False)
arg2015 = arg2015.drop(arg2015.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 

arg2016 = pd.read_csv('result_2016.csv',index_col=False)
arg2016 = arg2016.drop(arg2016.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 

arg2017 = pd.read_csv('result_2017.csv',index_col=False)
arg2017 = arg2017.drop(arg2017.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 

arg2018 = pd.read_csv('result_2018.csv',index_col=False)
arg2018 = arg2018.drop(arg2018.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 

arg2019 = pd.read_csv('result_2019.csv',index_col=False)
arg2019 = arg2019.drop(arg2019.columns[0], axis=1).sort_values(by = ['SQLDATEE']) 
```

```python
arg = pd.concat([arg2014,arg2015,arg2016,arg2017,arg2018,arg2019])
orignical_data = arg.copy()
```

```python
orignical_data.head()
```

```python
df = arg[['SQLDATEE','GoldsteinScale','NumMentions','NumSources','NumArticles','AvgTone']].reset_index(drop=True)
df.columns = ['DATE','GoldsteinScale','NumMentions','NumSources','NumArticles','AvgTone']
```

```python
agg1 = df.groupby('DATE').mean().reset_index()
```

```python
agg2 = df[['AvgTone','DATE']].groupby('DATE').count().rename(columns = {'AvgTone':'NumNews'}).reset_index()
```

```python
data = pd.merge(agg1, agg2, on='DATE')
```

```python
data.head()
```

```python
l = data['NumNews'].values.tolist()
Target = ['',]
for i,j in zip(range(0,len(l)-1), range(1,len(l))):
    if l[i] < l[j]:
        Target.append(1)
    else:
        Target.append(-1)
```

```python
Target = pd.DataFrame({'Target':Target})
```

```python
final_data = pd.concat([data,Target],axis=1)
```
